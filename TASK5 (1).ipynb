```python
zip_path = 'C:/Users/Priya Dutta/Downloads/titanic.zip' 
dfs = []
```

```python
with zf.ZipFile('C:/Users/Priya Dutta/Downloads/titanic.zip', 'r') as z:
    for file_name in ['train.csv', 'test.csv', 'gender_submission.csv']:
        with z.open(file_name) as f:
            df = pd.read_csv(f)
            dfs.append(df)
```

```python
combined_df = pd.concat(dfs, ignore_index=True)
print(combined_df.head())
```

```python
combined_df.to_csv('combined_titanic.csv', index=False)
```
